Now!! we need to impute this missing values or drop them if needed from df1 datframe.

```python
cols= ['AMT_REQ_CREDIT_BUREAU_YEAR','AMT_REQ_CREDIT_BUREAU_QRT', 'AMT_REQ_CREDIT_BUREAU_MON', 'AMT_REQ_CREDIT_BUREAU_WEEK','AMT_REQ_CREDIT_BUREAU_DAY','AMT_REQ_CREDIT_BUREAU_HOUR']
fig,axes = plt.subplots(ncols=3,nrows=2,figsize=(15,15))
count=0
for i in range(0,2):
    for j in range(0,3):
        sns.boxenplot(y=df1[cols[count]],ax=axes[i,j])
        count+=1
plt.show()
```

#From this plots we can see there are lots of outliers present in these columns, so to fill the missing values we treat them by bringing out the modes of 'AMT_REQ_CREDIT_BUREAU_YEAR','AMT_REQ_CREDIT_BUREAU_QRT', 'AMT_REQ_CREDIT_BUREAU_MON', 'AMT_REQ_CREDIT_BUREAU_WEEK','AMT_REQ_CREDIT_BUREAU_DAY','AMT_REQ_CREDIT_BUREAU_HOUR' columns..

```python
# process to impute the missing values in above mentioned columns....
for i in empty1.index.to_list():
    if 'AMT_REQ_CREDIT' in i:
        print('The mode value  {0} is : {1}'.format(i,df1[i].mode()[0]))
        df1[i].fillna(df1[i].mode()[0],inplace=True)
        print('NULL Values in {0} after imputation : {1}'.format(i,df1[i].isnull().sum()))
        print()
```

We have imputed the missing values for "AMT_REQ_CREDI" type of columns with mode values.

```python
def plot_merge(appdata_merge, column_name):
    col_value = ['Refused', 'Approved', 'Canceled', 'Unused offer']
    fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 12))
    axes = axes.flatten()

    for i, value in enumerate(col_value):
        if column_name!= 'PRODUCT_COMBINATION':
            data_subset = appdata_merge[appdata_merge['NAME_CONTRACT_STATUS'] == value]
            sns.countplot(data=data_subset, x=column_name, ax=axes[i])
            axes[i].set_title(value)
            axes[i].set_xlabel(column_name)
            axes[i].set_ylabel('Count')
            axes[i].tick_params(axis='x', rotation=45)  # Rotate x-axis labels by 45 degrees
        else:
            data_subset = appdata_merge[appdata_merge['NAME_CONTRACT_STATUS'] == value]
            sns.countplot(data=data_subset, x=column_name, ax=axes[i])
            axes[i].set_title(value)
            axes[i].set_xlabel(column_name)
            axes[i].set_ylabel('Count')
            axes[i].tick_params(axis='x', rotation=90)  # Rotate x-axis labels by 90 degrees
            

    plt.tight_layout()
    fig.suptitle("NAME_CONTRACT_STATUS VS " + column_name, fontsize=16, y=1.05)  # Adjust the y-coordinate of the suptitle

    plt.show()
```

```python
def plot_pie_merge(appdata_merge, column_name):
    col_value = ['Refused', 'Approved', 'Canceled', 'Unused offer']
    fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 12))
    axes = axes.flatten()

    for i, value in enumerate(col_value):
        data_subset = appdata_merge[appdata_merge['NAME_CONTRACT_STATUS'] == value]
        counts = data_subset[column_name].value_counts()
        labels = counts.index
        sizes = counts.values
        axes[i].pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
        axes[i].set_title(value)

    plt.tight_layout()
    fig.suptitle("NAME_CONTRACT_STATUS VS " + column_name, fontsize=16, y=1.05)
    plt.show()
```
